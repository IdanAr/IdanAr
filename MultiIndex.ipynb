# Create a MultiIndex on a DataFrame with the set_index Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"])
bigmac.head(3)
```

```python
bigmac.set_index(keys = ["Date", "Country"], inplace =True)   # Now there are 2 indexes. we can define how many indexes we want to.

# As we can see, it's like a PIVOT TABLE in Excel.
# So in learning, we should try to take the one with the least of values, and put it first. ( It's like an aggregator)
#So if i am going to have a MULTIINDEX which of my columns should I put in the outermost layer Vs. the innermost layer.
# The genral best practice is to take the column with the least number of UNIQUE values and put it on the most outer lvl
# and continue that realtionship as I proceed downwards.
```

```python
bigmac.sort_index(inplace = True )   # Sorting the INDEX, first it sort the DATE (which is first in line) and then COUNTRY (second).
bigmac
```

# Change Index Level Name with the set_names Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace =True)
bigmac.head(3)
```

```python
# In order to set a new name fgor the index label, we will use set_names() method.

bigmac.index.set_names(names = ["Day", "Location"],inplace =True)
```

```python
bigmac.head(3)
```

```python
# For another excrise, let's change back the Day index (previously "Date")
# to "Date", but now we will do it just on that column:
bigmac.index.set_names(names = "Date", level = 0, inplace = True)
bigmac.head(3)
```

# The sort_index Method on a MultiIndex DataFrame

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.head(3)
```

```python
bigmac.sort_index(ascending = False)  # Rearranging in descnding order.
# In the next code we will see how to set an order in multiindex in diffrent orders.
# The first column (Date) will be in ascending order, the second (Country) will be in descnding
bigmac.sort_index(ascending = [True, False])
```

```python
# Another way to sort index is with the argument of level.
# if we decided to sort only one index column, we use this syntax:
bigmac.sort_index(level = 1)  # We can see that the output is sorted by country column.
```

# Extract Rows from a MultiIndex DataFrame

```python
# Using iloc accessor:
bigmac.iloc[0]
```

```python
bigmac.iloc[[10,20,340]]
```

# The .swaplevel() Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace =True)
bigmac.head(3)
```

```python
bigmac.swaplevel()   # IF we have 2 indexes as multiindexes, there is no need to define 
                     # something in the (). pandas will undestand to swipe between the 2 of them
```

```python
# Lets assume we have 3 multiindexes. in the () we need to spacify just 2 index (No matter
# which one first because it will just swipe them)

bigmac.swaplevel(0,1)
bigmac.swaplevel(1,0)

bigmac.swaplevel("Date","Country")
bigmac.swaplevel("Country", "Date")

# The code above will do the same.
```

# The .stack() Method

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
world.stack()  # stack function make the columns as rows, which means it will make the lines (rows) bigger
```

```python
world.stack().to_frame()
```

#  The .unstack() Method, Part 1

```python
# unstacked function do the oppsite of stack. which means it take the most inner ROWS
# and transfer them into columns.
# We can call unstack as many times as the dataframe stack, until there is no rows. (It's erase the multiIndex)
s = world.stack()
s.unstack()  # The Population and GDP returned into a column.
```

```python
s.unstack().unstack() # Will take the year and put it as a column
```

# The .unstack() Method, Part 2

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
s= world.stack()
s.head(3)
```

```python
s.unstack(level = 0)  #Here it will take the index position of 0 (=country) and move it to the columns.
                      # At the default stat (.unstack()) its will take the last index (the most inner index)
```

# The .unstack() Method, Part 3

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
s = world.stack()
s.head(3)
```

```python
s.unstack(level = ["country", "year"])   # unstack more than 1 row using square brackets []
```

```python
s.unstack("year",fill_value = "NONE")  # fill value atgument let us to put anything in place of nulls values.
```

# The pivot Method

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sales.head(3)
```

```python
sales["Salesman"].value_counts()   # we have a lot of lines, but only 5 salesmen. 
                                   # convert to catagory.
```

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales.pivot(index = "Date", columns = "Salesman", values = "Revenue")

#as we can see in the output we made a pivot, using Dates as index
# Salesman as agregator (columns) and the values of revenue.
```

```python
salesman_pivot = sales.pivot(index = "Salesman", columns = "Date", values = "Revenue")
salesman_pivot.sort_index(axis = 0, ascending =False)
```

# Use the pivot_table method to create an aggregate summary of a DataFrame

```python
foods = pd.read_csv("foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "mean")  # aggfunc can be used
                                                                         # in many ways:
                                                                         # sum, min, max etc..
```

```python
foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "sum") 
```

```python
foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "max") 
```

```python
foods.pivot_table(values = "Spend", index = "Item", aggfunc = "mean") 
```

```python
# We can use MultiIndex here by addressing a few columns in index
foods.pivot_table(values = "Spend", index = ["Gender","Item"], aggfunc = "sum") 
```

```python
# Providing column(s) to the DataFrame Pivot Table:
foods.pivot_table(values = "Spend", index = ["Gender","Item"], columns = "City", aggfunc = "sum") 
```

```python
foods.pivot_table(values = "Spend", index = ["Gender","Item"], columns = ["Frequency","City"], aggfunc = "sum").head(10)
```

```python
# Another way to make a piuvot table is to call it from pandas library as shown below: (here we calling to a data set)
pd.pivot_table(data = foods,values = "Spend", index =
               ["Gender","Item"], columns = ["Frequency","City"], aggfunc = "count")
```
