# The pd.concat Method, Part 1

```python
# This method concat a two or more tables. 
# Here, we have variable of week1 & week2 which contains the same shape of columns and 
# respresent the same thing in diffrent week. we woult like to concat them.
# First, to concat, we have to use DF, so lets begin:
pd.concat(objs = [week1, week2], ignore_index = True)   # by default, ignore_index is False
                                                        # which brings week1 and week2 with
                                                        # the same index. If True it'll 
                                                        # extract week2 diffrent indexes.

```

```python
# Another way 2 concat Data-Frames is with append:
week1.append( other = week2, ignore_index = True)
```

# Inner Joins, Part 1

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
food = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.merge(week2, how = 'inner',on = "Customer ID")  #The output below show the customer
                                                      # which came on week1 and week2.
# At the result below, we can see that customer 155 have two rows. thats because he came to
# Resturant in week1 twice (2) and in week2 once (1) so python cannot decide where to merge
# these 2 ambigous result.
```

```python
# changing the suffixes (_x , _y)
week1.merge(week2, how = 'inner',on = "Customer ID", suffixes = [" - Week 1", " - Week 2"])
```

# Outer Join

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
food = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week 1", " - Week 2"])

# the outer brings all the result that ocour in one of the data sets.
# as we can see there are also null values in week2 and in week1. 
```

```python
# another cool thing is to use indicator = True. this method will show us from where the data
# came from.

```

```python
week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week 1", " - Week 2"], 
           indicator = True)
```

```python
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week 1", " - Week 2"], 
           indicator = True)
```

```python
merged["_merge"].value_counts()
```

```python
# If we want only the rows that occour in week1 or week2 but not on both of them:
mask = merged["_merge"].isin(["left_only", "right_only"])  # Crate boolean value.
```

```python
merged[mask]  
```

# Left Joins

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
food = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head(3)
```

```python
food.head(3)
```

```python
# if we will merge week1 and food (in left join method), it will show us just the rows from 
# week1 as Food ID that occour in Foods DB.
week1.merge(food, how = 'left', on = "Food ID")

# another thing we can put on merge function, is a sort method.
# this method will take the column that we merged on, and sort it in a way.
week1.merge(food, how = 'left', on = "Food ID", sort = True)
```

# The left_on and right_on Parameters

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
food = pd.read_csv("Restaurant - Foods.csv")
```

```python
#If we don't have the same column name for example:
week2.head(3)
```

```python
customers.head(3)
```

```python
# Here we have 2 databases and we want to merge it on week2.Customer ID and customers.ID col
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID" ).head(3)
```

```python
# As we can see, there 2 columns with the same values ("Customer ID" = "ID")
# We can drop one of them using "Drop" Method.
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID", 
            sort = True).drop("ID", axis = 1).head(3)
```

# Merging by Indexes with the left_index and right_index Parameters

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv", index_col = "ID")
food = pd.read_csv("Restaurant - Foods.csv", index_col = "Food ID")

# We changed a bit the indexes of 2 DB. The indexes are now on the ID of these DB.
```

```python
customers.head(3)
```

```python
food.head(3)
```

```python
# Because we changed the databases indexes, we need to spacify with a new parameter:
# right_index, in this case.
week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True)
```

```python
# Now we store this output to "sales" variable and to another merge.
sales = week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True)
```

```python
sales.merge(food, how = "left", left_on = "Food ID", right_index = True)
```

```python
# Now we will overwrite sales variable again to the last output we got:
sales = sales.merge(food, how = "left", left_on = "Food ID", right_index = True)
sales.head(3)
```
