# Filter a DataFrame Based on A Condition

```python
df = pd.read_csv("employees.csv", parse_dates= ["Start Date", "Last Login Time"])  #df = DataFrame, Dates modifing. 
df["Senior Management"] = df["Senior Management"].astype("boolean")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.head(3)
```

```python
df["Gender"] == 'Male'  # Will return boolean (True / False)
```

```python
df[df["Gender"] == 'Male']  # will return all the values in which we wanted to filter.(Only MALES)
```

```python
df[df["Team"] == "Finance"]  # will return all the values in which we wanted to filter.(Only FINANCE)
```

```python
mask = df["Team"] == "Finance"  # Another way to do that is to define a VARIABLE
                                # and then to call it through the DataFrame
df[mask]
```

```python
# If we have a boolean variable, we cannot use the same syntex because when we will
# make a comparison, every record will be True because False equal False and True equal True --> It will bring us True
# as long as we have a boolean series of any type as we have here in the "Senior Management" column ,
# we can pass it directly in our square brackets and pandas will return only the rows where there is a TRUE value.
df[df["Senior Management"]]
```

```python
mask = df["Team"] != "Marketing" # Not Equal (To "Marketing")
df[mask]
```

```python
df[df["Salary"] >110000]
# OR
mask = df["Salary"] > 110000
df[mask]
```

```python
df[df["Bonus %"] < 1.5]
# OR
mask = df["Bonus %"] < 1.5
df[mask]
```

```python
mask = df["Start Date"] <= "1985-01-01"
df[mask]
```

# Filter DataFrame with More than One Condition (AND - &)

```python
df = pd.read_csv("employees.csv", parse_dates= ["Start Date", "Last Login Time"])  #df = DataFrame, Dates modifing. 
df["Senior Management"] = df["Senior Management"].astype("boolean")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.head(3)
```

```python
# Gender = 'Male', Team = "Marketing"
df["Gender"] == "Male"
# Now we Store it in variable named "mask1"
mask1 = df["Gender"] == "Male"

df["Team"] == "Marketing"
# Now we Store it in variable named "mask2"
mask2 = df["Team"] == "Marketing"

# Merging the 2 conditions
df[mask1 & mask2].head()
```

# Filter DataFrame with More than One Condition (OR - |)

```python
df = pd.read_csv("employees.csv", parse_dates= ["Start Date", "Last Login Time"])  #df = DataFrame, Dates modifing. 
df["Senior Management"] = df["Senior Management"].astype("boolean")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.head(3)
```

```python
# Now we Store "Senior Management" in variable named "mask1"
mask1 = df["Senior Management"]

# Now we Store "Start Date" in variable named "mask2"
mask2 = df["Start Date"] < "1990-01-01"

# Merging the 2 condition into OR scenarios:
df[mask1 | mask2].head()
```

```python
# Here we learn how to use "AND" & "OR" in the same time.
mask1 = df["First Name"] =="Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3]  # Using paranthsis (Regualar) to define the filtered scenario.
```

# Check For Inclusion Within a Range of Values with the BETWEEN Method

```python
df = pd.read_csv("employees.csv", parse_dates= ["Start Date", "Last Login Time"])  #df = DataFrame, Dates modifing. 
df["Senior Management"] = df["Senior Management"].astype("boolean")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.head(3)
```

```python
df["Salary"].between(60000,70000) # it will return boolean.
df[df["Salary"].between(60000,70000)] # it will return the schema

#also we can use a variable as usual:
salary = df["Salary"].between(60000,70000)
df[salary]
```

```python
# Also when we talk about time (like clock) we can use BETWEEN method. Pandas will understand the contex.
# Last Login Time between 08:30 to 12:00 
time = df["Last Login Time"].between("08:30AM", "12:00PM")
df[time]
```

# Check for Duplicate DataFrame Rows with the duplicated Method

```python
df = pd.read_csv("employees.csv", parse_dates= ["Start Date", "Last Login Time"])  #df = DataFrame, Dates modifing. 
df["Senior Management"] = df["Senior Management"].astype("boolean")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
df[df["First Name"].duplicated()]  #Output: all the duplicated ones.
```

```python
df[df["First Name"].duplicated(keep = 'last')]  #this scenario will eliminate the last one that duplicated
df[df["First Name"].duplicated(keep = 'first')] #this scenario will eliminate the first one that duplicated
```

```python
df[df["First Name"].duplicated(keep = False)] # it will return *ALL* the duplicated, no matter if it's the first or last)
```

```python
df["First Name"].duplicated(keep = False) # This one is tricky. We want to see the UNIQUE ones.
                                          # well in this case we need to use ~ sign, which provides the opposite
                                          # of duplicated.
df[~df["First Name"].duplicated(keep = False)]
```
